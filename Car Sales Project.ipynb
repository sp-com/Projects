### DETECTING ANOMALIES

```python
sns.boxplot(df['Fuel_efficiency'])

plt.rcParams['figure.figsize'] = (5,10)

plt.show
```

```python
Q1 = df['Fuel_efficiency'].quantile(0.25) #checking the outliers for fuel efficiency
Q3 = df['Fuel_efficiency'].quantile(0.75)

IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
```

```python
IQR
```

### ELIMINATE OUTLIERS

```python
df_no_outliers = df[(df['Fuel_efficiency'] >= lower) & (df['Fuel_efficiency'] >= upper)]
print(f"Removed {df.shape[0] - df_no_outliers.shape[0]} outliers")
df_clean = df_no_outliers.copy()
```
